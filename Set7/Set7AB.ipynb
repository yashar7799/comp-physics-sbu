```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('stockindex.xlsx')
```

```python
data = df['Findex-Close'].values[1000:]
```

```python
n_terms = 4

coefficients = np.fft.fft(data)

coefficients[n_terms+1:-n_terms] = 0

smoothed_data = np.fft.ifft(coefficients).real
```

```python
plt.plot(data)
plt.plot(smoothed_data)
```
